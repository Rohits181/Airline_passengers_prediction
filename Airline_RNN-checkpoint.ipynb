```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Set default plot size and resolution
plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['figure.dpi'] = 100
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU
```

```python
# 🔹 Step 1: Load dataset
Data = pd.read_csv("AirPassengers.csv")
df = Data[['#Passengers']].copy() if '#Passengers' in Data.columns else Data.iloc[:, 1:].copy()
df.columns = ['Passengers']
```

```python
# 🔹 Step 2: Normalize data
scaler = MinMaxScaler()
df['Passengers'] = scaler.fit_transform(df)
```

```python
# 🔹 Step 3: Visualize the time series
df.plot(title='Monthly International Airline Passengers')
plt.ylabel("Normalized Passengers")
plt.grid(True)
plt.show()
```

```python

#  Step 4: Create sequences
def create_sequences(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)
```

```python
#  Step 5: Prepare data for RNN
def prepare_data(df, window_size, train_split=0.8):
    data = df.values
    X, y = create_sequences(data, window_size)
    split_idx = int(len(X) * train_split)

    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]

    X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

    return X_train, X_test, y_train, y_test
```

```python
#  Step 6: Build the RNN model
def build_model(model_type, input_shape):
    model = Sequential()
    if model_type == 'SimpleRNN':
        model.add(SimpleRNN(50, activation='tanh', input_shape=input_shape))
    elif model_type == 'LSTM':
        model.add(LSTM(50, activation='tanh', input_shape=input_shape))
    elif model_type == 'GRU':
        model.add(GRU(50, activation='tanh', input_shape=input_shape))
    else:
        raise ValueError("Invalid model type")
    
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
# 🔹 Step 7: Train, evaluate, and visualize

def train_and_evaluate(model_type, window_size, plot=True):
    X_train, X_test, y_train, y_test = prepare_data(df, window_size)
    model = build_model(model_type, input_shape=(window_size, 1))
    history = model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=16,
        validation_split=0.2,
        verbose=0
    )
    y_pred = model.predict(X_test)
    rmse = sqrt(mean_squared_error(y_test, y_pred))
    print(f'{model_type} (Window Size = {window_size}) RMSE: {rmse:.4f}')
    if plot:
        plt.figure(figsize=(10, 4))
        plt.plot(y_test, label='Actual')
        plt.plot(y_pred, label='Predicted')
        plt.title(f'{model_type} Predictions (Window={window_size})')
        plt.legend()
        plt.show()

        plt.figure(figsize=(10, 4))
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title(f'{model_type} Loss Curve (Window={window_size})')
        plt.legend()
        plt.show()

    return rmse, y_test, y_pred, history
```

```python
# 🔹 Step 8: Run the models for different window sizes
results = []
for window_size in [5, 10, 15]:
    for model_type in ['SimpleRNN', 'LSTM', 'GRU']:
        print(f"\nEvaluating {model_type} with window size {window_size}...")
        rmse, y_test, y_pred, history = train_and_evaluate(model_type, window_size)
        results.append((model_type, window_size, rmse))
```
